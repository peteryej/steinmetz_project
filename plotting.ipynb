```python
session_n = 24

X_transformed_ica_go_vis = np.array(data[str(session_n)][0])
X_transformed_ica_nogo_vis = np.array(data[str(session_n)][1])
X_transformed_ica_go_midbrain = np.array(data[str(session_n)][2])
X_transformed_ica_nogo_midbrain = np.array(data[str(session_n)][3])

X_transformed_ica_stimulus_vis = np.array(data[str(session_n)][4])
X_transformed_ica_nostimulus_vis = np.array(data[str(session_n)][5])
X_transformed_ica_stimulus_midbrain = np.array(data[str(session_n)][6])
X_transformed_ica_nostimulus_midbrain = np.array(data[str(session_n)][7])

print(X_transformed_ica_go_vis.shape)

# print(X_transformed_ica_nogo_vis.shape)
```

```python
%matplotlib notebook
# %matplotlib inline
# plotting 3D

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def plot_3d_traces(fig, ax, X_transformed_ica_left, X_transformed_ica_right, title='dummy title', left_label='left_label', right_label='right_label', left_color='m', right_color='c'):  

    ax.plot(X_transformed_ica_left[:,0], X_transformed_ica_left[:,1], X_transformed_ica_left[:,2], c=left_color, linewidth=2, label=left_label)
    ax.plot(X_transformed_ica_right[:,0], X_transformed_ica_right[:,1], X_transformed_ica_right[:,2], c=right_color, linewidth=2, label=right_label)

    N_points = X_transformed_ica_left.shape[0]
    marker_points_inx = [0, 50, 130, N_points-1]

#     ax.scatter(X_transformed_ica_left[marker_points_inx,0], X_transformed_ica_left[marker_points_inx,1],X_transformed_ica_left[marker_points_inx,2], c=['g', 'b','y', 'r'], linewidth=4)
#     ax.scatter(X_transformed_ica_right[marker_points_inx,0], X_transformed_ica_right[marker_points_inx,1],X_transformed_ica_right[marker_points_inx,2], c=['g', 'b','y', 'r'], linewidth=4)

    ax.set_title(title)


    ax.set_xlim(0.0, 0.01)
    ax.set_ylim(-0.001, 0.0014)
    ax.set_zlim(-0.003, 0.008)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.grid(False)

    ax.set_xlabel('Component 1')                
    ax.set_ylabel('Component 2')
    ax.set_zlabel('Component 3')

    # Set the angle of the camera
    ax.view_init(20, 45)

    plt.legend()
    plt.tight_layout()
    plt.show()
    # plt.savefig(f'3D_trajectories_go.svg')


```

```python

fig = plt.figure(figsize=(9,8))
ax = fig.add_subplot(221, projection='3d')
plot_3d_traces(fig, ax, X_transformed_ica_go_vis[30:70] , X_transformed_ica_nogo_vis[30:70], title='Region: vis ctx', left_label='go', right_label='no-go', left_color='b', right_color='g')

ax = fig.add_subplot(222, projection='3d')
plot_3d_traces(fig, ax, X_transformed_ica_stimulus_vis[30:70], X_transformed_ica_nostimulus_vis[30:70], title='Region: vis ctx', left_label='right-side stimulus', right_label='no right-side stimulus')

ax = fig.add_subplot(223, projection='3d')
plot_3d_traces(fig, ax, X_transformed_ica_go_midbrain[110:150], X_transformed_ica_nogo_midbrain[110:150], title='Region: midbrain',  left_label='go', right_label='no-go', left_color='b', right_color='g')

ax = fig.add_subplot(224, projection='3d')
plot_3d_traces(fig, ax, X_transformed_ica_stimulus_midbrain[110:150], X_transformed_ica_nostimulus_midbrain[110:150], title='Region: midbrain', left_label='right-side stimulus', right_label='no right-side stimulus')

fig.suptitle('ICA Reduced 3 Components for Session 24')
plt.tight_layout()
plt.savefig(f'3D_trajectories_2_brain_regions.png')
```
